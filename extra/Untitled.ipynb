```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.font_manager as fm

# Path to the font on your computer
font_path = '/Library/Fonts/SF-Pro-Display-Regular.otf'

# Register the font
font_prop = fm.FontProperties(fname=font_path)
plt.rcParams['font.family'] = font_prop.get_name()

# Data
datasets = ["BEHAVE", "DTCPS", "ESM2", "ESPLE", "GAMESEFINAL", "LC", "MPM4CPS", "RL4SE",
            "SECSELFADAPT", "SMELLREPROD", "TESTNN", "TRUSTSE", "UPDATECOLLABMDE"]
ws_scores = [92, 84, 89, 88, 86, 72, 72, 95, 77, 96, 62, 66, 91]
bacc_scores = [90, 72, 61, 60, 72, 65, 81, 79, 58, 87, 56, 63, 71]

# DataFrame
df = pd.DataFrame({
    'Dataset': datasets,
    'Work Saved': ws_scores,
    'Balanced Accuracy': bacc_scores
})

# Create scatter plot (A1 is ~23.4 × 33.1 inches, using your provided size for convenience)
plt.figure(figsize=(10.2, 6.7))  

# Scatter plot points
plt.scatter(df['Work Saved'], df['Balanced Accuracy'], color='blue', s=120, marker='x')

# Add dataset labels carefully to avoid overlapping
label_offsets = {
    "BEHAVE": (0.5, 0.5),
    "DTCPS": (0.5, -1),
    "ESM2": (0.5, -1.5),
    "ESPLE": (0.5, -1),
    "GAMESEFINAL": (0.5, 0.5),
    "LC": (0.5, 0.5),
    "MPM4CPS": (0.5, 0.5),
    "RL4SE": (0.5, -1),
    "SECSELFADAPT": (0.5, 0.5),
    "SMELLREPROD": (-7.5, 0.5),
    "TESTNN": (0.5, 0.5),
    "TRUSTSE": (0.5, -1.5),
    "UPDATECOLLABMDE": (-10, -1.5)
}

for i, row in df.iterrows():
    offset = label_offsets.get(row['Dataset'], (0.5, 0.5))
    plt.text(row['Work Saved'] + offset[0], 
             row['Balanced Accuracy'] + offset[1],
             row['Dataset'],
             fontsize=13,
             ha='left', va='bottom')

# Set titles and axis labels
plt.title('Scatter Plot of Work Saved vs Balanced Accuracy Scores', fontsize=16, pad=20)
plt.xlabel('Work Saved (%)', fontsize=14)
plt.ylabel('Balanced Accuracy (%)', fontsize=14)

# Set axis ticks
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Set axis limits to start from 50%
plt.xlim(50, 100)
plt.ylim(50, 100)

# Add grid
plt.grid(True, linestyle='--', alpha=0.7)

# Save the plot as high-resolution SVG
output_svg = '/Users/gauransh/scatter_plot_A1_high_res.svg'
plt.savefig(output_svg, format='svg', dpi=1200, bbox_inches='tight')
plt.close()

print(f"Scatter plot saved successfully at {output_svg}")
```
